```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier,AdaBoostClassifier, GradientBoostingClassifier
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from sklearn.ensemble import RandomForestClassifier
from sklearn.utils.class_weight import compute_class_weight
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
url = "https://raw.githubusercontent.com/Mancini997/mini_project_ML_wine_and_smile/main/WineQuality.csv"
df = pd.read_csv(url)
```

```python
df= df.drop(columns=['Unnamed: 0'])
```

```python
print(df['quality'].value_counts())
```

```python
df_num = df.select_dtypes(include=['number'])
```

```python
target= df_num["quality"]
```

```python
features= df_num.drop(columns =['quality', ])
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size = 0.20, random_state=0)
```

```python
X_train.head()
```

```python
y_train.value_counts()
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(X_train, y_train)
```

```python
knn.score(X_test, y_test)
```

```python
df_num.hist(bins=20, figsize=(10, 10))
plt.show()
```

```python
normalizer = MinMaxScaler()
```

```python
normalizer.fit(X_train)
```

```python
X_train_norm = normalizer.transform(X_train)

X_test_norm = normalizer.transform(X_test)
```

```python
X_train_norm
```

```python
X_train_norm = pd.DataFrame(X_train_norm, columns = X_train.columns)
X_train_norm.head()
```

```python
knn.fit(X_train_norm, y_train)
```

```python
knn.score(X_test_norm, y_test)
```

```python
X_train_norm.hist(bins=20, figsize=(10, 10))
plt.show()
```

```python
#Random Forest and Bagging
```

```python
bagging_class = BaggingClassifier(DecisionTreeClassifier(max_depth=30),
                               n_estimators=200,
                               max_samples = 2500)
```

```python
bagging_class.fit(X_train_norm, y_train)
```

```python
pred = bagging_class.predict(X_test_norm)

print("MAE", mean_absolute_error(pred, y_test))
print("RMSE", mean_squared_error(pred, y_test, squared=False))
print("R2 score", bagging_class.score(X_test_norm, y_test))
```

```python
# Calculate class weights based on class distribution
class_counts = {
    3: 133,
    4: 1144,
    5: 10755,
    6: 14117,
    7: 5371,
    8: 941,
    9: 24
}

# Total number of instances
total_samples = sum(class_counts.values())

# Compute the class weights
class_weight_dict = {cls: total_samples / (len(class_counts) * count) for cls, count in class_counts.items()}

print("Class Weights: ", class_weight_dict)

# Initialize the RandomForestClassifier with the computed class weights
rforest = RandomForestClassifier(class_weight=class_weight_dict, random_state=42, n_estimators=100)

# Train the model on the normalized training data
rforest.fit(X_train_norm, y_train)

# Predict the test set
y_pred = rforest.predict(X_test_norm)

# Evaluate the model on the test set
print("MAE:", mean_absolute_error(y_pred, y_test))
print("RMSE:", mean_squared_error(y_pred, y_test, squared=False))
print("R2 score:", rforest.score(X_test_norm, y_test))

```

```python
def classify_input(model, feature_names):  
    user_input = []

    for feature in feature_names:
        value = input(f"Enter the value for {feature}: ")
        value = value.replace(',', '.')
        user_input.append(float(value))

    user_input = np.array(user_input).reshape(1, -1)
    prediction = bagging_class.predict(user_input)
    
    return prediction[0]
```

```python

feature_names = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', ' total sulfur dioxide',	'density',	'pH',	'sulphates','alcohol']
prediction = classify_input(bagging_class.predict, feature_names)
print(f"The Selected Wine is Quality: {prediction} on a scale 3 to 9")
```
