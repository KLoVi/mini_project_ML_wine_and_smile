```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/KLoVi/mini_project_ML_wine_and_smile/main/WineQuality.csv"
```

```python
df = pd.read_csv(url)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
df.head(3)
```

```python
df.rename( columns = { "Unnamed: 0" : "wine_instances"} , inplace = True )
```

```python
df.head(3)
```

```python
df.shape
```

```python
df = pd.get_dummies(df, columns=["Type"], drop_first=True) #,drop_first=True  
```

```python
df
```

```python
# White Wine == TRUE ; Red Wine == FALSE
```

```python
df["Type_White Wine"] = df["Type_White Wine"].astype(int)
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df_num = df.select_dtypes( include = ["number"] )
```

```python
df_num  = df_num.drop(columns = ["wine_instances"])
```

```python
df_num.head(3)
```

```python
df_num.describe().T.round(2)
```

```python
target = df_num["quality"]
```

```python
features = df_num.drop(columns = ["quality"])
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.20, random_state=0)
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(X_train, y_train)
```

```python
pred = knn.predict(X_test)
pred
```

```python
y_test.values
```

```python
knn.score(X_test, y_test)
```

```python
correlation_matrix = df_num.corr()

# Setting up the matplotlib figure with an appropriate size
plt.figure(figsize=(10, 7))

# Drawing the heatmap for the numerical columns
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", annot_kws={"size": 9})


plt.title("Correlation Heatmap for Selected Numerical Variables")
plt.show()
```

```python
df_num.hist(bins=20, figsize=(10, 10))
plt.show()
```

```python
# FEATURE ENGINEER
```

```python
normalizer = MinMaxScaler()
```

```python
normalizer.fit(X_train)
```

```python
X_train_norm = normalizer.transform(X_train)

X_test_norm = normalizer.transform(X_test)
```

```python
X_train_norm = pd.DataFrame(X_train_norm, columns = X_train.columns)
X_train_norm.head()
```

```python
X_test_norm = pd.DataFrame(X_test_norm, columns = X_test.columns)
X_test_norm.head()
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(X_train_norm, y_train)
```

```python
knn.score(X_test_norm, y_test)  #The result doesn't change if Red or White wine are converted into binary.
```
